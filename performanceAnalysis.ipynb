```python
# Plot outside of notebook
%matplotlib qt

from matplotlib import cm

# Load the results.npy file
results = np.load("results.npy", allow_pickle=True)

bh = results[:,:,0].T
bf = results[:,:,1].T

hf = plt.figure()
ax1 = hf.add_subplot(121, projection='3d')
ax2 = hf.add_subplot(122, projection='3d')

def plllt(ax, data):
  X, Y = np.meshgrid(range(1,17), range(1000, 11000, 1000))

  # Plot time per particle
  data *= X/Y

  ax.plot_surface(X,Y,data)
  ax.contour(X, Y, data, zdir='z', offset=0, cmap='coolwarm')
  ax.contour(X, Y, data, zdir='x', offset=0, cmap='coolwarm')
  ax.contour(X, Y, data, zdir='y', offset=0, cmap='coolwarm')

plllt(ax1, bh)
plllt(ax2, bf)
```
