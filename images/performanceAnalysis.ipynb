```python
import matplotlib.pyplot as plt
import numpy as np

cores = np.array(range(1,17))
```

```python
# Load the results.npy file
results = np.load("results.npy", allow_pickle=True)

bh = results[:,:,0].T
bf = results[:,:,1].T

sls = 9
plt.plot(cores, bh[sls,0]/bh[sls,:])
plt.plot(cores, bf[sls,0]/bf[sls,:])

plt.ylabel("x times speedup")
plt.xlabel("Number of cores")
plt.title("Speedup for gravity simulation with 10,000 particles")
plt.legend(["barnes hut", "brute force"])
```

```python
# Plot outside of notebook
%matplotlib qt

from matplotlib import cm

hf = plt.figure()
ax1 = hf.add_subplot(121, projection='3d')
ax2 = hf.add_subplot(122, projection='3d')

def plllt(ax, inputData):
  X, Y = np.meshgrid(cores, range(1000, 11000, 1000))

  # Plot time per particle
  data = inputData.copy() * X/Y

  ax.plot_surface(X,Y,data)
  ax.contour(X, Y, data, zdir='z', offset=0, cmap='coolwarm')
  ax.contour(X, Y, data, zdir='x', offset=0, cmap='coolwarm')
  ax.contour(X, Y, data, zdir='y', offset=0, cmap='coolwarm')

plllt(ax1, bh)
plllt(ax2, bf)
```
